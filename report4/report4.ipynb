```julia
#問１
using Plots 
using NLsolve

function problem(u)
    return sin(u)-3/4
end


function nls(func, params...; ini = [0.0])
    if typeof(ini) <: Number
        r = nlsolve((vout,vin)->vout[1]=func(vin[1],params...), [ini])
        v = r.zero[1]
    else
        r = nlsolve((vout,vin)->vout .= func(vin,params...), ini)
        v = r.zero
    end
    return v, r.f_converged
end


#二分法

MAX = 100
top = π/2
bottom = 0
err  = 1e-8
solve = 0

for i in 1:MAX
    temp = (bottom + top)/2
    
    if (problem(temp) > 0) 
        top = temp
    elseif (problem(temp) == 0)
        solve = temp
        #たまたま寸分違わぬ解を当てることはほぼ確実にありえないけどないと言い切るのも良くない気がする。
        break
    else
        bottom = temp
    end
    
    if abs(top - bottom) < err
        break
    end
    #所定の誤差以下で離脱。
    
    
    print("$i : 下限：$bottom　　上限：$top\n")
end
solve = (bottom + top)/2 

print("近似解：$solve\n")



#ソルバー

sol = nls(problem, ini = 1.0)
print("NLsolveによる解：$sol")




```

```julia
#問２

using LinearAlgebra
using Plots
using NLsolve

function nls(func, params...; ini = [0.0])
    if typeof(ini) <: Number
        r = nlsolve((vout,vin)->vout[1]=func(vin[1],params...), [ini])
        v = r.zero[1]
    else
        r = nlsolve((vout,vin)->vout .= func(vin,params...), ini)
        v = r.zero
    end
    return v, r.f_converged
end


MAX = 100 
#ニュートン法の試行回数上限
err = 1e-7
#ニュートン法の誤差上限


function prob(v)
    x, y, z = v
    return [
        x+2*y+z+1
        x^2+2*y^2+z^2-10
        sin(x+y+z)-0.7
    ]
end
#解くべき非線形方程式


function probdiv(v)
    x,y,z = v
    return[
        1 2 1
        2*x 4*y 2*z
        cos(x+y+z) cos(x+y+z) cos(x+y+z)
    ]
end
#上の非線形方程式を微分した行列

#二分法は初期値を決めるのが少し難しそうなので、ニュートン法とNlsolveだけで行く。
ini_v = [1.0,0.0,0.0]
temp = ini_v
solve = [0,0,0]
print(prob(temp))
print(probdiv(temp))

for i in 1:MAX
    
    print("$i : 近似値$temp ,誤差$(norm(prob(temp)))\n")
    if norm(prob(temp)) < err
        solve = temp
        break
    end
    
    temp = temp - probdiv(temp)\prob(temp)
end

print("近似解:$solve\n")

sol = nls(problem , ini = ini_v)
print("NLsolveによる近似解:$sol")
```
