```julia
#問１
using Plots 
using NLsolve

function problem(u)
    return sin(u)-3/4
end


function nls(func, params...; ini = [0.0])
    if typeof(ini) <: Number
        r = nlsolve((vout,vin)->vout[1]=func(vin[1],params...), [ini])
        v = r.zero[1]
    else
        r = nlsolve((vout,vin)->vout .= func(vin,params...), ini)
        v = r.zero
    end
    return v, r.f_converged
end


#二分法

MAX = 100
top = 0
bottom = π/2
err  = 1e-8
solve = 0

for i in 1:MAX
    temp = (bottom + top)/2
    
    if (problem(temp) > 0) 
        top = temp
    elseif (problem(temp) == 0)
        solve = temp
        #たまたま寸分違わぬ解を当てることはほぼ確実にありえないけどないと言い切るのも良くない気がする。
        break
    else
        bottom = temp
    end
    
    if abs(top - bottom) < err
        break
    end
    #所定の誤差以下で離脱。
    
    
    print("下限：$bottom　　上限：$top\n")
end

print("近似解：$solve\n")



#ソルバー

sol = nls(problem, ini = 1.0)
print("NLsolveによる解：$sol")




```
