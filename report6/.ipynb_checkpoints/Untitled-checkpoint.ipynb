```julia
#文字に関してはあとでGoogleDocumentを使って作ること。

using Plots 


γ = 1
#ここもちょっといじってみたい。スケールが変わるだけだろうか？
u0 = 0.1
#初期値は授業資料と同じ。
TIME = 10

function aitken(x,n)
    return x[n] - (x[n+1]-x[n])^2 / (x[n+2] - 2*x[n+1] + x[n])
end

#Aitken加速では、n番目の計算のためにn+2番目までの値を使う。なので、計算範囲は数列の大きさより２つ小さくする必要がある。

function DP(r,u,Δt)
    r1 = r(u)*Δt
    r2 = r(u + (1/5)*r1 )*Δt
    r3 = r(u + (3/40)*r1 + (9/40)*r2 )*Δt
    r4 = r(u + (44/45)*r1 - (56/15)*r2 + (32/9) *r3 )*Δt
    r5 = r(u + (19372/6561) *r1 - (25360/2187) *r2 + (64448/6561) *r3 - (212/729) *r4 )*Δt
    r6 = r(u + (9017/3168) *r1 - (355/33) *r2 + (46732/5247) *r3 + (49/176) *r4 - (5103/18656) *r5 )*Δt

    return u + (35/384)*r1 + (500/1113)*r3 + (125/192)*r4 - (2187/6784)*r5 + (11/84)*r6
end

#引数は微分方程式の右辺。引数には関数rと値uを与えること。
#この部分は授業資料からコピペ。

function problem(u)
    return γ*(sin(u))^1.2
end

function accel(sq)
    #aitken加速をする関数を先に作っておく。これなら、適当な回数加速する操作が簡単になるし、範囲指定を間違うこともない。
    ac = [aitken(sq,i) for i in 1:length(sq)-2]
    return ac
end



#解きたい微分方程式の右辺に当たる関数。左辺はuのt微分になる。
#t = 10の時の値について調べることにする。(TIME = 10)

sol_sq = []
delt_sq = []

for n in 200:1:500
    #TIMEまでの範囲をn個に分割する。
    delt = TIME/n
    push!(delt_sq,delt)
    #Δtの値の記録をとっておく。
    
    temp = u0
    for i in 1:1:n
        temp = DP(problem,temp,delt)
    end
    
    push!(sol_sq , temp)
    #計算結果を収納。
end


plt = plot(delt_sq,sol_sq , markershape = :circle, ls = :solid)
xlabel!(plt,"Δt [s]")
ylabel!(plt,"Calculated Value at t =10")
title!(plt,"Δt - Values at t=10")
savefig(plt,"delt_sol.png")
display(plt)

#作った結果をaitken加速で更に収束させる。
#気になったので何回かけるかでどう違うのか表示させて確かめて見る。


once = accel(sol_sq)
twice = accel(once)
thrice = accel(twice)
fourth = accel(thrice)

ax = plot(once,label ="Acceled once")
plot!(ax,twice,label = "Acceled twice")
plot!(ax,thrice,label = "Acceled thrice")
plot!(ax,fourth,label = "Acceled fourth")
ylabel!("value at t =10")
savefig(ax,"aitken_second.png")

display(ax)

print(fourth)


#メモ
#ルンゲクッタ法による解法自体はうまく行っており、計算値も妥当。
#aitken加速をかけた時に、うまく行かないことが多い。NANやINFなどが入ってくる。すべてがそうではないので加速そのものが間違っているわけではなさそう。
#分母になる部分が小さすぎて計算がうまく行っていないのではないか。




```

```julia
using Plots 


γ = 1.0
TIME = 10
u0 = 0.1

function problem(u,t=1 )
    return γ*(sin(u))^1.2
end


    
u_ex = 3.1184213196531

function DP(r,u,Δt)
    r1 = r(u)*Δt
    r2 = r(u + (1/5)*r1 )*Δt
    r3 = r(u + (3/40)*r1 + (9/40)*r2)*Δt
    r4 = r(u + (44/45)*r1 - (56/15)*r2 + (32/9) *r3 )*Δt
    r5 = r(u + (19372/6561) *r1 - (25360/2187) *r2 + (64448/6561) *r3 - (212/729) *r4 )*Δt
    r6 = r(u + (9017/3168) *r1 - (355/33) *r2 + (46732/5247) *r3 + (49/176) *r4 - (5103/18656) *r5 )*Δt

    return u + (35/384)*r1 + (500/1113)*r3 + (125/192)*r4 - (2187/6784)*r5 + (11/84)*r6
end


function RKF(r,u,t,delt)
    r1 = r(u,t)
    r2 = r(u + r1*delt/4 , t+delt/4)
    r3 = r(u + delt*(3*r1/32 + 9*r2/32), t + 3*delt/8)
    r4 = r(u + delt*(1932*r1/2197 -7200*r2/2197 + 7296*r3/2197) , t + 12*delt/13)
    r5 = r(u + delt*(439*r1/216 -8*r2 + 3680*r3/513 + 845*r4/4104) , t+delt)
    r6 = r(u + delt*(-8*r1/27 + 2*r2 - 3544*r3/2565 + 1859*r4/4104 -11*r5/40) , t+delt/2)
    
    four = u + delt*(25*r1/216 + 1408*r3/2565 + 2197*r4/4104 -r5/5)
    five = u + delt*(16*r1/135 + 6656*r3/12825 + 28561*r4/56430 -9*r5/50 + 2*r6/55)
    return four , five
end


function ab(x1 , y1 ,x2,y2)
    a = (y2-y1)/(x2-x1)
    b = y1 - a * x1
    return a,b
end


#それぞれの計算方法による結果を収納する配列を準備しておく。
DPsolve = []
RKFsolve = []
RKFfivesolve = []
delt_sq = []

#あとでこれらから一気に厳密回を引いて誤差を求めることにする。

for n in 100:500
    delt = TIME/n
    push!(delt_sq , delt)
    
    dp = u0
    rk = u0
    waste = 0
    for i in 1:n
        dp = DP(problem , dp , delt)
        rk , waste = RKF(problem , rk ,i*delt, delt)
    end
    push!(DPsolve , dp)
    push!(RKFsolve , rk)
    #それぞれの方法で計算してあとはそれを配列に格納するだけ。
end


DPerr = [log10(abs(k - u_ex)) for k in DPsolve]
RKFerr = [log10(abs(k - u_ex)) for k in RKFsolve]
delt_log = [log10(d) for d in delt_sq]

plt = plot(delt_log , DPerr , markershape = :circle  , label = "Dormand Prince method")
plot!(plt,delt_log, [5.476138*d - 5.321079 for d in delt_log],label = "Approximate line")
ylabel!(plt, "error in log")
xlabel!(plt, "Δt in log")
savefig("DMmethod_error.png")
display(plt)

DPa, DPb = ab(delt_log[1],DPerr[1] , delt_log[30],DPerr[30])
print("傾き: a = $DPa , 切片: b = $DPb")

ax = plot(delt_log , RKFerr,markershape = :circle, label = "Runge- kutta-Fehlberg method")
plot!(ax,delt_log , [1.01841 * d - 2.76757 for d in delt_log] , label = "Approximate line")
xlabel!(ax,"Δt in log")
ylabel!(ax, "error in log")
savefig("RKFmethod_error.png")
display(ax)

RKFa,RKFb = ab(delt_log[100] , RKFerr[100] , delt_log[300] , RKFerr[300])
print("傾き: a = $RKFa , 切片: b = $RKFb")

typeof(DPerr[100])
```
